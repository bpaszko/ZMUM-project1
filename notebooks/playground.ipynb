```python
# Correlations
df_num = df.drop(non_number_cols, axis=1).drop('class', axis=1)
corrs = []
for name in df_num:
    corrs.append((name, df_num[name].corr(df['class'])))
    
for name, corr in sorted(corrs, key=lambda x: x[1]):
    print(f'{name}: {corr}')
```
